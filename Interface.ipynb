```python
from tkinter import *
from spellchecker import SpellChecker
from tkinter import messagebox
from textAnalyzer import textAnalyzer
import numpy as np

spell = SpellChecker()
root = Tk()
root.title("Text Analyzer")
canvas = Canvas(root,width=498, height=400, bg='grey')
canvas.pack(expand=YES, fill=BOTH)   
heading = Label(canvas,text = "TextAnalyzer", bg = "grey",fg = "black",font = 'Courier 25 bold')
heading.place(x = 130, y = 20)
root.resizable(height=False,width=False)


def autocorrect(word):
    spell = SpellChecker()
    correctWord = spell.correction(word)
    return correctWord

global prevChange
#prevChange = 0


def correct(event):
    global unknown_text
    global prev_text
    global flag
    global prevChange
    global text

    text = entry.get(1.0,"end-1c")
    if(len(str(text).strip())>0):
        flag = 0
        prev_text = text
        text = str(text).split(" ")
        unknown_text = text[-1]
        isdot = unknown_text[-1]
        #unknown_text = list(spell.unknown(text))[-1]
        if(len(text)==1):
            text = [text[0].capitalize()]
        
        if(len(text)>2 and text[-2][-1] == "."):
            text[-1] = text[-1].capitalize()


        if (len(unknown_text)>0 and prevChange==0):
            #for word in unknown_text:
            correctWord = spell.correction(unknown_text)
            for index, value in enumerate(text):
                if(unknown_text == value):
                    text[index] = correctWord
                    if(isdot == "."):
                        text[-1] = str(text[-1]) + "."
                    flag = 1
            
            text = " ".join(text)
            entry.delete("1.0", 'end')
            entry.insert("1.0",text)
        prevChange = 0
    

def prevSent(event):
    global prevChange
    global flag
    prevChange = 0   
    new_text = entry.get(1.0,"end-1c")
    if(len(str(new_text).strip())>0):
        new_text = new_text.split()
        new_text[-1] = unknown_text
        new_text = " ".join(new_text)
        if(len(new_text)>len(text)):
            pass
        else:
            if(flag == 1):
                entry.delete("1.0", END)
                entry.insert("1.0",str(new_text)+" ")
                prevChange = 1
                flag = 0
prev_text = ""


def analyse():
    x1 = 100
    y1 = 50
    y2 = 50      
    x = 510
    y = 75 
    x2 = 0
    global Emotionpred
    global data
    Emotionpred = None
    data = entry.get(1.0,"end-1c")
    if(len(data)>0):
        canvas.config(width = 800, height = 400)
        heading.place(x = 280, y = 20 )
        analyzer = textAnalyzer(data)
        Emotionpred = analyzer.textEmotion(textProcess=False)
        sentiment = analyzer.textSentiment(textProcess=True)
        Emotionpred = Emotionpred + sentiment
        for index in range(len(Emotionpred)):
            x2 = Emotionpred[index][1]*100+120
            # if(index == 6):
            #     Label(canvas,text = ["-"*20]).place(x = 250,y = 100)
            if(Emotionpred[index][1]>0):
                line = canvas.create_line(x1, y1, x2, y2, width=20, fill='green')    
                canvas.move(line, 510, 35)
            else:
                pass

            catLabel = Label(canvas,text = Emotionpred[index][0],bg = "grey").place(x = x, y= y) #Category Label
            per = np.round(Emotionpred[index][1]*100,2)
            per_text = str(per)+"%"
            percent = Label(canvas,text = per_text , bg = "grey").place(x = x+240, y= y) #Percentage
            
            y1 += 30
            y2 += 30   
            y +=30
    else:
         messagebox.showerror("Error","Enter Sentence")
         canvas.config(width=498, height=400)

def clear():
    
    canvas.delete("line")
    canvas.delete("all")
    
    entry.delete("1.0", END)
    heading.place(x = 130, y = 20)
    canvas.config(width=498, height=400,bg = "grey")


entry = Text(root,height = 15, width = 61)
entry.place(x = 5,y = 70)
#canvas.create_window(250,200,height = 250, width = 495,window=entry)
entry.bind('<space>',correct)
entry.bind('<BackSpace>',prevSent)

button = Button(canvas,text = "Analyse",command=analyse)
button.place(x =130, y = 350,width=100)
button.bind('<Return>',analyse)
button = Button(canvas,text = "Clear",command=clear)
button.place(x =260, y = 350, width = 100)



mainloop()
```
